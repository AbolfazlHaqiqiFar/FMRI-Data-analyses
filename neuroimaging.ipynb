```python
import nilearn as nl
import nibabel as nib
from nilearn import datasets
import seaborn as sns
import pandas as pd
import scipy.ndimage as ndi
import numpy as np
import matplotlib.colors as colors
import nilearn.image as nlimg
import nilearn.plotting as nlplt
```

```python
path_to_file = "melodic_IC_sum.nii.gz"

# Load the Nifti image using Nilearn
nifti_img = nl.image.load_img(path_to_file)

# Print some information about the image
print("Shape of the image data:", nifti_img.shape)
print("Voxel dimensions:", nifti_img.header.get_zooms())
```

```python
for i in range(5):
    vol1_img = nlimg.index_img(nifti_img, i)

    # Plot the 3D image using Nilearn
    plot = nlplt.plot_anat(vol1_img, cmap='magma', colorbar=True, annotate=True)
    #plot.savefig(f"image{i}.jpg")
```
